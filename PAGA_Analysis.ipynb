```python
import anndata
import scanpy as sc

data = anndata.read_h5ad("batch_integrated.h5ad")
```

```python
sc.pl.umap(data, color=["cell_type_detail"],legend_loc='on data', show=True)
```

```python
sc.pl.umap(data, color=["dataset"], show=True)

```

```python
sc.pl.umap(data, color=["batch"], show=True)
```

```python
fcd_OD = data[data.obs['batch'].isin(['townsend_JLE49_3', 'westfall_JLE18', 'westfall_JLE18'])]
fcd = data[data.obs['batch'].isin(['townsend_JLE48_3', 'townsend_JLE50_3'])]
control = data[data.obs['batch']=="westfall_NB1499"]


#sc.pl.umap(fcd, color=["leiden"], legend_loc = "on data", show=True)
#sc.pl.umap(control, color=["leiden"], legend_loc = "on data", show=True)
```

```python
sc.tl.paga(fcd_OD, groups = 'cell_type_detail')

# visualize the PAGA graph
sc.pl.paga(fcd_OD,node_size_scale=3, edge_width_scale=.5)

```

```python
sc.tl.paga(fcd, groups = 'cell_type_detail')

# visualize the PAGA graph
sc.pl.paga(fcd,node_size_scale=3, edge_width_scale=.5)

```

```python
sc.tl.paga(control, groups = 'cell_type_detail')

# visualize the PAGA graph
sc.pl.paga(control,node_size_scale=3, edge_width_scale=.5)
```

```python
import scanpy as sc
import matplotlib.pyplot as plt


# Create a figure with multiple subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))



# Plot PAGA for each dataset
sc.pl.paga(fcd_OD, ax=axs[0])
axs[0].set_title('Dataset 1')
sc.pl.paga(fcd, ax=axs[1])
axs[1].set_title('Dataset 2')
sc.pl.paga(control, ax=axs[2])
axs[2].set_title('Dataset 3')

# Show the plot
plt.show()
```
